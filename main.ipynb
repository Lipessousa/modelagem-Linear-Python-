    <h2><center>Modelagem Linear aplicada para previsão de lucro e prejuízo em uma transação comercial<center/></h2>
    <p>
    Fazemos parte do time de Sistemas de Informação de uma grande multinacional do setor farmacêutico que está realizando as primeiras transações comerciais de um composto orgânico inovador que trará uma grande revolução no setor de medicamentos anti-inflamatórios. Entretanto, por ser um setor altamente competitivo, os dados são altamente sigilosos e, mesmo dentro da própria corporação, tanto a ordem de grandeza dos preços de custo e receita quanto a forma de mensurar a quantidade produzida não são declaradas. Entretanto, os dados fornecidos (mesmo numa codificação de sigilo) são suficientes para analisar as possibilidades de lucro e prejuízo.
    <p/>
    <p>
        Nosso setor recebeu alguns dados de custos de produção e de receitas de venda como função da quantidade produzida do composto. Esses dados foram gerados a partir dessas primeiras transações comerciais realizadas. Nossa missão é modelar esses dados e entregar o mais rápido possível as previsões de lucro e prejuízo. Devemos ser bastante precisos e meticulosos nessa análise pois qualquer erro pode ser financeiramente fatal para o lançamento em escala mundial desse produto. 
    <p/>
<div/>


<div style="border: 1px solid #fff; padding: 10px">
    <p>
    Seja x a quantidade produzida de um certo produto. O custo total de produção (ou simplesmente custo) depende de x, e a relação entre eles chamamos de função custo total (ou simplesmente função custo), e a indicamos por <font color="blue">C</font>. Existem custos que não dependem da quantidade produzida, tais como aluguel, seguros e outros. A soma desses custos que não dependem da quantidade produzida chamamos de custo fixo e indicamos por C_F. A parcela do custo que depende de x chamamos de custo variável, e indicamos por <font color="blue">C_V</font>. Assim, podemos escrever: <font color="blue">C=C_F+C_V.</font>Verificamos também que, para x variando dentro de certos limites (normalmente não muito grandes), o custo variável é geralmente igual a uma constante multiplicada pela quantidade x. Essa constante é chamada de custo variável por unidade. Por exemplo, se o custo fixo mensal de fabricação de um produto é $ 5000,00 e o custo variável por unidade é $ 10,00, então a função custo total será dada por <font color="blue">C(x)=5000+10x</font>.
    <p/>
    <p>
        Seja x a quantidade vendida de um produto. Chamamos de função receita ao produto de x pelo preço de venda e a indicamos por R. Por exemplo, se um produto é vendido a $ 15,00 a unidade (preço constante), então a função receita será <font color="blue">R(x)=15x</font>.
    <p/>
    <p>
        A função lucro é definida como a diferença entre a função receita R(x) e a função custo C(x). Assim, indicando a função lucro por <font color="blue">L</font>, teremos <font color="blue">L(x)=R(x)-C(x)</font>.
    <p/>
<div/>

    A tabela a seguir apresenta os dados de custos de produção C(x) e de receitas de venda R(x) como função da quantidade produzida x do composto que foram gerados das primeiras transações comerciais. Apresente os gráficos de custos de produção como função da quantidade produzida e das receitas de venda como função da quantidade produzida.
<p/>

<p>
    Sendo as colunas <font color="blue">Qtde. produzida(x)</font> e <font color="blue">Custo Total(Y) </font>fazendo parte de Custos de Produção.

<p>
    E as colunas <font color="blue">Qtde. Vendidas(x)<font/> e <font color="blue">Receita(y)<font/> fazendo parte de Receitas de Venda.

```python
import pandas as pd
import numpy as np

table = pd.read_csv('base_dados_custo_venda.csv')
display(table)
```

```python
import matplotlib.pyplot as plt

#Dados de custo de produção
x_prod_qtd = table['Qtde. produzida (x)']
y_prod_cost = table['Custo Total (y)']

#Aqui é ajustado um polinomio do primeiro grau aos dados, retornando os coeficientes
prod_coefficient_linear = np.polyfit(x_prod_qtd, y_prod_cost, 1)
print("coeficientes lineares dos Dados de Produção: ", prod_coefficient_linear)

#Cria um função polinomial do primero grau a partir dos coeficientes
prod_linear_equation = np.poly1d(prod_coefficient_linear)
print("equação que corresponde a função Custo: ", prod_linear_equation)


```

<p  style="border-bottom: 1px solid #fff; display: flex; justify-content: center">
    Neste caso 2 é o valor de custo variável, ja que é o valor que é multiplicado pela quantidade X, fazendo com que o valor de custo fixo seja 50
</p>

```python
#Dados de receita de venda
x_sell_qtd = table['Qtde. vendida (x)']
y_sell_revenue = table['Receita (y)']

#Aqui é ajustado um polinomio do primeiro grau aos dados, retornando os coeficientes
sell_coefficient_linear = np.polyfit(x_sell_qtd, y_sell_revenue, 1)
print("coeficientes lineares dos Dados de Venda: ", sell_coefficient_linear)

#Cria um função polinomial do primero grau a partir dos coeficientes
sell_linear_equation = np.poly1d(sell_coefficient_linear)
print("equação que corresponde a função Receita: ", sell_linear_equation)
```

<p  style="border-bottom: 1px solid #fff; display: flex; justify-content: center">
    Neste caso 4 é o valor de custo variável, ja que é o valor que é multiplicado pela quantidade X
</p>

```python
#Variáveis para estilização do Gráfico
fig, ax = plt.subplots()

#Estilização do gráficos
ax.set_xlabel('Qtde. Produzida / Vendida')
ax.set_ylabel('Custo Total / Receita')
ax.set_title('Custo de Produção X Receitas de Venda')
ax.grid(True, color="grey")

#Plotagem dos dados de Produção
plt.scatter(x_prod_qtd, y_prod_cost, label="Custo de Produção", color="blue", edgecolors="black")
#Plotagem dos dados de Venda
plt.scatter(x_sell_qtd, y_sell_revenue, label="Receita de Venda", color="green", edgecolors="black")

#Cria as retas lineares no gráfico
plt.plot(x_prod_qtd, y_prod_cost, color='black')
plt.plot(x_sell_qtd, y_sell_revenue, color='black')


plt.legend()
plt.show()
```

<div style="border: 1px solid #fff">
<h3><center>Construção da Função Lucro<center/></h3>

<p><center>C(x)=2x+50<center><p/>
<p><center>L(x)=R(x)-C(x)<center><p/>
<p><center>L(x)=4x-(2x+50)<center><p/>
</div>

```python
#Criação da Função Lucro
profit_equation = sell_linear_equation - prod_linear_equation

print('Equação que correpsonde a Função Lucro:', profit_equation)
```

```python
#Variável que contém o cáculo do lucro
profit = y_sell_revenue - y_prod_cost

#Variáveis para estilização do Gráfico
fig, ax = plt.subplots()

#Estilização do gráficos
ax.set_xlabel('Qtde. Produzida')
ax.set_ylabel('Lucro / Prejuízo')
ax.grid(True, color="black")

#Plotagem dos dados de Produção
plt.scatter(y_prod_cost, profit, label="Custo de Produção", color="black", edgecolors="red")

#Cria reta linear no gráfico
plt.plot(y_prod_cost, profit, color="black")

plt.legend()
plt.show()
```
